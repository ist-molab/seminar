```python
import numpy as np
import math
import matplotlib.pyplot as plt
from matplotlib import cm
import statistics as st
from scipy import stats
%matplotlib inline

plt.rcParams["font.size"] = 18

bin_size = 100
num_trial = 100000
sample_size = [1,5,10,30,50,100]
```

## 中心極限定理における分布収束の定数倍の重さについて考える

## 一様分布$U(0,1)$(歪度=0)

```python
fig = plt.figure(figsize = (20, 12), dpi = 200)


for i, n in enumerate(sample_size):
    ax = fig.add_subplot(2, 3, i+1)
    plt.title(f'number of sum ={n}')
    mu = stats.uniform.mean()
    var = stats.uniform.var()

    Xs = [np.mean(stats.uniform.rvs(size = n)) for _ in range(num_trial)]
    Xs = [(p - mu)*n**(1/2)/var**(1/2) for p in Xs]   
    
    
    plt.hist(Xs, bins = bin_size,density = True)
    x = np.linspace(-6,6)
    norm_vals = stats.norm.pdf(x)
    
    plt.plot(x, norm_vals, label = 'norm')
    plt.legend(fontsize = 12)
plt.subplots_adjust(left = 0.05, right=0.95, bottom = 0.05, top = 0.95)
plt.show()
```

## 指数分布$exp(1)$(歪度=2)

```python
fig = plt.figure(figsize = (20, 12), dpi = 200)


for i, n in enumerate(sample_size):
    ax = fig.add_subplot(2, 3, i+1)
    plt.title(f'number of sum ={n}')
    mu = stats.expon.mean()
    var = stats.expon.var()

    Xs = [np.mean(stats.expon.rvs(size = n)) for _ in range(num_trial)]
    Xs = [(p - mu)*n**(1/2)/var**(1/2) for p in Xs]   
    
    
    plt.hist(Xs, bins = bin_size,density = True)
    x = np.linspace(-6,6)
    norm_vals = stats.norm.pdf(x)
    
    plt.plot(x, norm_vals, label = 'norm')
    plt.legend(fontsize = 12)
plt.subplots_adjust(left = 0.05, right=0.95, bottom = 0.05, top = 0.95)
plt.show()
```

## beta分布$beta(2,5)$

```python
# a,bの値で歪度可変
a = 0.1
b = 10

print("歪度 = ", 2*(b-a)*math.sqrt(a+b+1)/(a+b+2)/math.sqrt(a*b))
```

```python
fig = plt.figure(figsize = (20, 12), dpi = 200)


for i, n in enumerate(sample_size):
    ax = fig.add_subplot(2, 3, i+1)
    plt.title(f'number of sum ={n}')
    mu = stats.beta.mean(a,b)
    var = stats.beta.var(a,b)

    Xs = [np.mean(stats.beta.rvs(a,b,size = n)) for _ in range(num_trial)]
    Xs = [(p - mu)*n**(1/2)/var**(1/2) for p in Xs]   
    
    
    plt.hist(Xs, bins = bin_size,density = True)
    x = np.linspace(-6,6)
    norm_vals = stats.norm.pdf(x)
    
    plt.plot(x, norm_vals, label = 'norm')
    plt.legend(fontsize = 12)
plt.subplots_adjust(left = 0.05, right=0.95, bottom = 0.05, top = 0.95)
plt.show()
```
